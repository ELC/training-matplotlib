# Necessary Imports

```python
from random import randint
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Bar Chart

```python
X = [1, 5, 8, 2, 4, 6, 8]
Y = [1, 8, 2, 4, 8, 9, 3]

plt.figure(figsize=(17, 9))
plt.bar(X, Y, label="Bar")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.legend()

plt.show()
```

## Multiple Bars Chart

```python
X = [1, 0, 8, 1, 4, 6, 8]
Y = [1, 8, 2, 4, 8, 9, 3]

X2 = [1, 3, 5, 9, 11]
Y2 = [8, 7, 1, 6, 1]

plt.figure(figsize=(17, 9))
plt.bar(X, Y, label="Bar1")
plt.bar(X2, Y2, label="Bar2")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.legend()

plt.show()
```

## Adding different Colours

```python
X = [1, 0, 8, 3, 4]
Y = [1, 8, 2, 4, 8]

X2 = [1, 0, 8, 3, 4]
Y2 = [9, 7, 1, 6, 1]

plt.figure(figsize=(17, 9))
plt.bar(X, Y, label="Bar1", color='r')
plt.bar(X2, Y2, label="Bar2", color='c')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.legend()

plt.show()
```

## Histogram vs Bar Charts

```python
number_of_points = 700
max_value = 20
min_value = 0
number_of_bins = 11

X = [i for i in range(number_of_points)]
Y = [randint(min_value, max_value) for _ in range(number_of_points)]

plt.figure(figsize=(17, 9))
plt.bar(X, Y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Bar Chart')
plt.show()

bins = np.arange(max_value + 2) - 0.5

plt.figure(figsize=(17, 9))
plt.hist(Y, bins, histtype='bar', rwidth=0.7)

plt.xlabel('x')
plt.xticks(bins[:-1] + 0.5)  # Center the ticks
plt.ylabel('y')
plt.title('Histogram')
plt.show()
```
