```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Values
slices = [7, 2, 2, 13]

# Labels
activities = ['sleeping', 'eating', 'working', 'playing']

plt.figure(figsize=(9, 9))
plt.pie(slices, labels=activities,)

plt.title('Pie Chart')

plt.show()
```

# Pie Chart Customized

```python
# Values
slices = [7, 2, 2, 13]

# Labels
activities = ['sleeping', 'eating', 'working', 'playing']

plt.figure(figsize=(9, 9))
plt.pie(slices,
        labels=activities,
        startangle=90,  # Angle
        shadow=True,  # Shadow
        explode=(0.05, 0.05, 0.05, 0.05),  # Separation
        autopct="%1.1f%%")  # Show %

plt.title('Pie Chart')

plt.show()
```
