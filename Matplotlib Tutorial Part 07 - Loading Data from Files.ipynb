# Necessary Imports

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

filename = "Matplotlib Tutorial Part 07 - Data"
```

## CSV Method

```python
with open(filename) as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    X, Y = zip(*plots)

X = [int(i) for i in X]
Y = [int(i) for i in Y]

plt.figure(figsize=(17, 9))
plt.plot(X, Y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data loaded with CSV Method')

# Pottling
plt.grid(True)
plt.show()
```

## Numpy Method

```python
X, Y = np.loadtxt(filename, delimiter=',', unpack=True)

plt.figure(figsize=(17, 9))
plt.plot(X, Y)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Data loaded with Numpy Method')

# Plotting
plt.grid(True)
plt.show()
```
