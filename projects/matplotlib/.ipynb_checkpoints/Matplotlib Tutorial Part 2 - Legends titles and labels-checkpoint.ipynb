# Matplotlib Tutorial Part 2: Legends titles and labels

### [Source](https://www.youtube.com/watch?v=aCULcv_IQYw)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
X = [1,2,3]
Y = [5,6,1]

plt.plot(X, Y)

plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Function\nOr Graph')

plt.show()
```

## Several Graph on the same box

```python
X = [1,2,3]
Y = [5,6,1]

X2 = [1,2,3]
Y2 = [1,2,3]

plt.plot(X, Y, label='Graph1')
plt.plot(X2, Y2, label='Graph2')

plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Function\nOr Graph')

plt.legend()

plt.show()
```
