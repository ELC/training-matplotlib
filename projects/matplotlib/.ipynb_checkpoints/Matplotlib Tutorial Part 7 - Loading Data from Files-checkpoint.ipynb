# Matplotlib Tutorial Part 7 - Loading Data from Files

```python
%matplotlib inline
import matplotlib.pyplot as plt
filename = "Matplotlib Tutorial Part 7 - Data"
```

## CSV Method

```python
import csv

X = []
Y = []

with open(filename) as csvfile:
    plots = csv.reader(csvfile,delimiter=',')
    for row in plots:
        X.append(int(row[0]))
        Y.append(int(row[1]))

plt.plot(X,Y, label="Loaded from file!")        
        
plt.xlabel('x')
plt.ylabel('y')
plt.title('Loaded Data')
plt.legend()

plt.show()
```

## Numpy Method

```python
import numpy as np

X, Y = np.loadtxt(filename,
                  delimiter=',',
                  unpack=True)

plt.plot(X,Y, label="Loaded from file!")        
        
plt.xlabel('x')
plt.ylabel('y')
plt.title('Loaded Data')
plt.legend()

plt.show()
```
