```python
%matplotlib inline
import matplotlib.pyplot as plt
```

## Plotting Points

```python
X = [1,2,3]
Y = [5,6,1]

plt.plot(X, Y)
plt.show()
```

## Plotting Math Functions

```python
def f(x):
    return x**2

X = [i for i in range(-5,6)]
Y = [f(x) for x in X]

plt.plot(X,Y)
plt.show()
```

## Plotting Random points

```python
from random import randint

amount_points = 30

X = [i for i in range(amount_points)]
Y = [randint(0,9) for i in range(amount_points)]

plt.plot(X,Y)
plt.show()
```
